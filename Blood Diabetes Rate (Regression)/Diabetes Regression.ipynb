```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
%matplotlib inline
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
diabetes.DESCR
```

```python
diabetes.feature_names
```

```python
len(diabetes.data)
```

```python
diabetes_data = pd.DataFrame(data = diabetes.data,columns = ['age', 'sex', 'bmi', 'bp', 's1','s2', 's3', 's4', 's5', 's6'])
target = diabetes.target
```

```python
# diabetes_data = pd.Series()
diabetes_data['Target'] = target
```

```python
diabetes_data.head()
```

```python
plt.plot(diabetes_data['s1'],diabetes_data['Target'], '*')
```

```python
diabetes.DESCR
```

```python
diabetes_data = diabetes_data.drop(labels=['age', 'sex', 's1', 's2', 's3'],axis=1)
```

```python
diabetes_data
```

```python
diabetes_data.isnull().values.any()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = diabetes_data.drop(labels=['Target'],axis=1)
y = diabetes_data['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
reg = LinearRegression()
```

```python
reg.fit(X=X_train,y=y_train)
```

```python
predictions = reg.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,predictions)
```
